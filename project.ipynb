#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
def normalize(group):
    return (group - group.min()) / (group.max() - group.min())
normalize_prices = ds2.groupby('level5')['price'].transform(normalize)
ds2['normalize_prices']=normalize_prices

# Plot histograms for normalized prices of each population
plt.figure(figsize=(10, 6))
for population, group in ds2.groupby('level5'):
    plt.hist(group['normalize_prices'], bins=20, alpha=0.5, label=population)

plt.title('Normalized Price Distribution for Each Population')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
# TODimport pandas as pd
import matplotlib.pyplot as plt


# Step 1: Get the list of unique towns in the 'level5' column (should include 'Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón')
towns = ds2['level5'].unique()

# Step 2: Create subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Scatter Plots of Price vs surface for Different Towns')

# Step 3: Plot each town's scatter plot in its own subplot
for i, town in enumerate(towns):
    # Filter the data for each town
    town_data = ds2[ds2['level5'] == town]
    
    # Get the row and column index for the subplot (axes)
    row = i // 2
    col = i % 2
    
    # Create scatter plot for each town
    axes[row, col].scatter(town_data['surface'], town_data['price'], alpha=0.6)
    axes[row, col].set_title(town)
    axes[row, col].set_xlabel('surface (m²)')
    axes[row, col].set_ylabel('Price (€)')

# Adjust the layout to prevent overlap
plt.tight_layout()
plt.subplots_adjust(top=0.88)  # Adjust title space

# Show the plots
plt.show()
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
import folium
import pandas as pd
from folium import CircleMarker

# Correct the formatting of latitude and longitude
ds2['latitude'] = ds2['latitude'].str.replace(',', '.', regex=False)
ds2['longitude'] = ds2['longitude'].str.replace(',', '.', regex=False)
ds2['latitude'] = pd.to_numeric(ds2['latitude'], errors='coerce')
ds2['longitude'] = pd.to_numeric(ds2['longitude'], errors='coerce')
# Step 2: Create a map centered around Madrid (adjust the coordinates if necessary)
madrid_map = folium.Map(location=[40.416775, -3.703790], zoom_start=11)  # Coordinates of Madrid
# Step 3: Add each point to the map, coloring by town
colors = {
    'Getafe': 'blue',
    'Alcorcón': 'green',
    'Fuenlabrada': 'red',
    'Leganés': 'purple'
}
 #Add circle markers to the map for each row in the filtered DataFrame
for idx, row in ds2.iterrows():
    population = row['level5']
    coord = (row['latitude'], row['longitude'])
    circle_marker = CircleMarker(
        location=coord,
        radius=5,
        color=colors.get(population, 'gray'),  # Fallback to 'gray' if population not in 'colors'
        fill=True,
        fill_color=colors.get(population, 'gray'),
        fill_opacity=0.6
    )
    circle_marker.add_to(madrid_map)
# Save the map to an HTML file
madrid_map.save('madrid_south_belt_map.html')
# Display the map
madrid_map

```
